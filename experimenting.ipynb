```python
from google import genai
from google.genai import types
from dotenv import load_dotenv

load_dotenv()

import os


client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))
```

```python
print("My files:")
for f in client.files.list():
    print(" ", f.display_name)
    print(" ", f.name)
    print()
```
